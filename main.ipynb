**Calculating Team Stength**

```python
#splitting the data into two parts: df_home and df_away
df_historical
```

```python
df_home = df_historical[['HomeTeam','HomeGoals' ,'AwayGoals']]
df_away = df_historical[['AwayTeam','HomeGoals' ,'AwayGoals']]
```

```python
# renaming the columns
df_home.columns = ['Team','GoalsScored','GoalsConceded']
df_away.columns = ['Team','GoalsConceded','GoalsScored']
```

```python
print(df_home)
print(df_away)
```

```python
# combining the two dataframes and calculating the mean
df_team_strength = pd.concat([df_home,df_away], ignore_index=True).groupby('Team').mean()
df_team_strength
##  calculating the mean to check how good or bad a team is.
```
